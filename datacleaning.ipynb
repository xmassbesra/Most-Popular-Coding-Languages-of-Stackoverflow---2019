```python
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
# imported all the necessary libraries.
```

```python
plt.style.use("fivethirtyeight")
# used the plotting style from the 'Matplotlib'
```

```python
data = pd.read_csv("survey_results_public.csv")
pd.set_option("display.max_column", 85)
# loaded the data from resource.
# displayed all the columns of the csv file.
```

```python
data.head()
# to get along with the data, took a quick look of from '.head()' function.
```

```python
ids = data['Respondent']
lang_responses = data["LanguageWorkedWith"].astype(str)
# made some necessary temporary variables for choosing out the correct sets of datas.
```

```python
language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))
# used counter to gether & count all the datas from the 'LanguageWorkedWith' Column, using for loop.
```

```python
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
# populated the datas to & counted number to the empty lists, using for loop.     
```

```python
languages.reverse()
popularity.reverse()
# give them a reverse style for visual asthetic.
```

```python
plt.barh(languages, popularity)
plt.title("Most Popular Languages of 2019 as per Stackoverflow")
plt.xlabel("Number of people who use")
plt.show()
# plotted the data using necessary plotting title and label,
```
